```python
import matplotlib.pyplot as plt
import numpy as np 
import json
import pandas as pd
```

```python
versions = ["cp_model_1.0.0.mzn",
        "cp_model_1.0.0_rotations.mzn",
        "cp_model_1.0.0_rotations_sym.mzn",
        "cp_model_1.0.0_sym.mzn",
        "cp_model_1.3.0.mzn",
        "cp_model_1.3.0_rotations.mzn",
        "cp_model_1.3.0_rotations_sym.mzn",
        "cp_model_1.3.0_sym.mzn"]
```

```python
def read_json(time_table):
    with open(time_table, "r") as f:
        time_json = json.load(f)
    return time_json
```

```python
times = []
for i in range(0,len(versions)): 
    time = read_json(versions[i].replace("mzn","json"))
    # if i == 0:
        # times.append(list(range(0,len(time.values()))))
    times.append(time.values())

# print(times)
```

```python
df = pd.DataFrame(times)
df = df.transpose()
df[df == -1] = 0
df.columns = versions
print(df)
```

```python
print("Mean time per instance:\n",df.mean(axis=0))
df.boxplot()
```
