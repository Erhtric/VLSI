```python
import matplotlib.pyplot as plt
import numpy as np 
import json
import pandas as pd
```

```python
no_rot_no_sym = "time_no_rot_no_sym.json"
no_rot_sym = "time_no_rot_sym.json"
rot_no_sym = "time_rot_no_sym.json"
rot_sym = "time_rot_sym.json"

versions = [no_rot_no_sym, \
            no_rot_sym,
            rot_no_sym,
            rot_sym]
```

```python
def read_json(time_table):
    with open(time_table, "r") as f:
        time_json = json.load(f)
    return time_json
```

```python
times = []
for i in range(0,len(versions)): 
    time = read_json(versions[i])
    # if i == 0:
        # times.append(list(range(0,len(time.values()))))
    times.append(time.values())

# print(times)
```

```python
df = pd.DataFrame(times)
df = df.transpose()
df.columns = versions
print(df)
```

```python
print("Mean time per instance:\n",df.mean(axis=0))
df.boxplot()
```
